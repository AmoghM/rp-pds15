## Week 6: Data Acquisition

This week we will start working with data by using Python to read,
write, parse, and a visualize data in different formats. We first review
different data formats and present standard Python mechanisms to read
and write data into these different formats. Next we review how to
actually parse specific data out of data formats that are not directly
transformable into Python data structures, like XML data. Next, we
introduce the Pandas library, which can greatly simplify many data
processing tasks. Finally, we present an example Notebook that combines
many of these capabilities to make an interesting data visualization,
known as a [Choropleth][clp] , seen below.

![Our IPython Notebook visualization](images/cusa.svg)

### Topic 1: [Data Formats][1]

This Notebook introduces many of the popular data formats including
fixed-width text, delimiter separated values, JSON, XML, and HDF. Example
code demonstrated how to read and write to all of these formats by using
only standard Python libraries, except for HDF.

### Topic 2: [Introduction to Data Parsing][2]

This Notebook introduces the concept of parsing a structured data format
in order to extract useful data. The Document Object Model is
introduced, which is generated by many XML parsers. The BeautifulSoup
library is introduced to simplify data parsing and is used to first
parse an HTML document, and later our own XML document.

### Topic 3: [Python Data Analysis][3]

This Notebook introduces the Pandas library, specifically focusing on
the `Series` and `DataFrame` data structures. Techniques for reading
data into a `DataFrame` from different types of data are also introduced
as well as a discusion of how the `DataFrame` can be used to simplify
database access from within a Python program.

### Topic 4: [Data Visualization Example][4]

An IPython Notebook that provides a worked example, based on a
visualization originally created by Nathan Yau from FlowingData, that
combines disparate data, Python parsing and analysis to create a
Chloropeth based on unemployment data maintained by the US government.
